Q1. What is MongoDB? Explain non-relational databases in short. In which scenarios it is preferred to use MongoDB over SQL databases?

A1. MongoDB is a popular NoSQL document-oriented database that uses a flexible schema design to store and manage data. Unlike traditional relational databases that use tables with fixed columns and rows, MongoDB stores data in flexible JSON-like documents with dynamic schemas, which allows for faster and more flexible data modeling and retrieval. 

Non-relational databases, also known as NoSQL databases, are a type of database management system that doesn't use the traditional tabular structure of relational databases. NoSQL databases use flexibleand scalable data models that can adopt to different types of data and workloads.

MongoDB is often prefered over SQL databases in scenarios where flexible data model, scalability, high availability, Real-time analytics, cloud-native applications. 

Q2. State and Explain the features of MongoDB.

A2. Here are several key features of MongoDB:

1-Documented-Oriented: It is a document-oriented database, which means that it stores data n JSON-like documents with dynamic schemas.

2- Scalability: It is designed to scale horizontally across multiple nodes, making it easy to handle large volumes of data and high traffic loads.

3- Indexing: It support various types of indexing to optimizing query performance, including single-field, compound, geospatial, and text indexing.

Q5. Explain how you can use the find() method to query the MongoDB database. Write a simple code to demonstrate this.

A5. It is used to query documents in a collection. It  returns a cursor that can be used to iterate over the documents that match the query criteria. 

import pymongo


client = pymongo.MongoClient("mongodb://localhost:27017/")


db = client["mydatabase"]


collection = db["mycollection"]


docs = [
    {"name": "John", "age": 35},
    {"name": "Jane", "age": 28},
    {"name": "Bob", "age": 42},
    {"name": "Alice", "age": 31},
    {"name": "Tom", "age": 25},
]
collection.insert_many(docs)


query = {"age": {"$gt": 30}}
docs = collection.find(query)


print("Matching documents:")
for doc in docs:
    print(doc)


Q6. Explain the sort() method. Give an example to demonstrate sorting in MongoDB.

A6. It is usedto sort the documents in a collection based on one or more fields. It takes one parameter, which is a dictionary that specifies the sorting order. The keys of dictionary represnt the fields to be sorted, and the values represent the sorting direction.

The sorting direction can be specified as either'-1' for ascending order or '-1' for descending order. By default, the'sort()' method to sort documents in a MongoDB collection.

import pymongo


client = pymongo.MongoClient("mongodb://localhost:27017/")


db = client["mydatabase"]


collection = db["mycollection"]


docs = [
    {"name": "John", "age": 35},
    {"name": "Jane", "age": 28},
    {"name": "Bob", "age": 42},
    {"name": "Alice", "age": 31},
    {"name": "Tom", "age": 25},
]
collection.insert_many(docs)


query = {"age": {"$gt": 30}}
docs = collection.find(query).sort("name", 1)

Q7. Explain why delete_one(), delete_many(), and drop() is used.

A7. (i) 'delete_one()': This method is used to delete a asingle document that matches the given filter criteria. If there are multiple documents that match the criteria, onlt the first one that is found will be deleted.

(ii) 'delete_many()': This method is used to delete all documents that match the given filter criteria. It deletes all the documents that match the criteria, not just the first one.

(iii) 'drop()': This mehod is used to delete an entire collection from the database.
